# Recommendation System Engine

### Data Preprocessing

Import dataset

```python
import pandas as pd

df = pd.read_excel('name_movie_df.xlsx', index_col=0)  
```

```python
df
```

Let's extract the title from the index

```python
df.reset_index(inplace=True)
df = df.rename(columns = {'movie_name':'title'})
df['title'] = df['title'].astype(str)
df
```

```python
user_feat = pd.read_excel('user_feat.xlsx', index_col=0)  
user_feat.reset_index(inplace=True)
user_feat
```

```python
user_feat = user_feat.drop(['movie_id'], axis=1)
df = pd.concat([df, user_feat], axis=1, ignore_index=True)
df
```

```python
df.columns = df.columns.map(str)
```

```python
df = df.rename(columns={"0": "title"})
df
```

```python
movie_df = df
movie_df = movie_df.drop(['title'], axis=1)
movie_df = movie_df.T
```

```python
movie_df
```

Computing the similarity matrix

Here what I used

https://stackoverflow.com/questions/37003272/how-to-compute-jaccard-similarity-from-a-pandas-dataframe

```python
from sklearn.metrics.pairwise import pairwise_distances
jac_sim = 1 - pairwise_distances(movie_df.T, metric = "hamming")
# optionally convert it to a DataFrame
jac_sim = pd.DataFrame(jac_sim, index=movie_df.columns, columns=movie_df.columns)
```

```python
jac_sim
```

```python
jac_sim.shape
```

```python
#Construct a reverse mapping of indices and movie titles, and drop duplicate titles, if any
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
indices
```

```python
# Function that takes in movie title as input and gives recommendatmovie_dfions 
def jaccard_recommender(title, jac_sim=jac_sim, df=df, indices=indices):
    # Obtain the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    # And convert it into a list of tuples as described above
    sim_scores = list(enumerate(jac_sim[idx]))

    # Sort the movies based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies. Ignore the first movie.
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['title'].iloc[movie_indices]
```

```python
user_movies = pd.read_excel('user_movie_names.xlsx', index_col=None)  
user_movies = user_movies.rename(columns = {'movie_name':'title'})
user_movies
```
