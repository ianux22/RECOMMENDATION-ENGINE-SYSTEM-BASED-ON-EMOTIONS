# Working on KM 11

Let's import the data

```python
import pandas as pd
import numpy as np


movie = pd.read_csv("movie_sent_clust.csv")
movie
```

Removing index

```python
movie = movie.iloc[:,1:]
```

Let's put in movie_sent the sentiments

```python
movie_sent = movie.iloc[:, 0:45] 
movie_sent
```

and a sepratae table with all the clusters

```python
clusters = movie.iloc[:, 45:] 
clusters
```

let's merge sentiments and KM11

```python
movie_sent['KM11'] = clusters.KM11   #substitute km6 with another cluster vairable to insert it
movie_sent
```

### We are merging the group 1 and 4 because both romantic. Label "4" becomes "1"

```python
movie_sent['KM11'] = movie_sent['KM11'].replace([4],1)
movie_sent.KM11.value_counts()
```

Let's replace the rest of the clusters in order to have a sequence from 0 to 8

```python
movie_sent['KM11'] = movie_sent['KM11'].replace([5],4)
movie_sent['KM11'] = movie_sent['KM11'].replace([6],5)
movie_sent['KM11'] = movie_sent['KM11'].replace([7],6)
movie_sent['KM11'] = movie_sent['KM11'].replace([8],7)
movie_sent['KM11'] = movie_sent['KM11'].replace([9],8)
movie_sent['KM11'] = movie_sent['KM11'].replace([10],9)
movie_sent.KM11.value_counts()
```

Now we are going to standardize variables

```python
movie_sent['KM11'] = movie_sent.KM11.astype('object')
```

```python
cold_df = movie_sent.select_dtypes(exclude=["object"])
cold_df
```

### standardization

```python
cold_df = cold_df.drop(columns=['movie_id'])
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# the scaler object (model)
scaler = StandardScaler()
# fit and transform the data
scaled_data = scaler.fit_transform(cold_df) 
```

```python
cold_df = pd.DataFrame(scaled_data)
```

```python
cold_df
```

```python
cold_df['KM11'] = movie_sent['KM11']
cold_df
```

### Let's put the cluster 2 in the test dataset

```python
df_test = cold_df[cold_df['KM11'] == 2] 
x_test = df_test.iloc[:, 0:44]
y_test = df_test.iloc[:, 44:]
```

### cold_df has all the other clusters

```python
cold_df = cold_df[cold_df['KM11'] != 2] 
cold_df
```

```python
X = cold_df.iloc[:, 0:44]
y = cold_df.iloc[:, 44:]
```

##### Let's use a third of the dataset as validation set

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
 x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=0.35, random_state=5)
y_train=y_train.astype('int')
y_val=y_val.astype('int')
y_test=y_test.astype('int')
```

```python
y_val.value_counts()
```

```python
y_train.value_counts()
```

##### NOw we are ready to implement a random forest model

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
# random forest model creation
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train.values.ravel())
# predictions
rfc_predict = rfc.predict(x_val)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
rfc_cv_score = cross_val_score(rfc, x_train, y_train.values.ravel(), cv=10, scoring='accuracy')
```

#### Results on validation set

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_val, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_val, rfc_predict))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```
